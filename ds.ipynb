```python
import pandas as pd
df = pd.DataFrame({"Age":[19,20,21,25,24],"Name":['Aniket','Pratik','None','Aarti','Shivani']});
print(df)
```

```python
df.isnull()
```

```python
df['tick']=df['Name'].fillna('5')
df['tick']
```

```python
mean_age = df["Age"].mean()
mean_age
```

```python
new_age = df["Age"].fillna('mean_age')
new_age
```

```python
df
```

```python
df["Age"] = df["Age"].fillna(mean_age)
df
```

```python
med = df['Age'].median()
med
```

```python
df['Median']=df['Age'].fillna(med)
df
```

```python
Mod=df['Age'].mode()
Mod
```

```python
df['Mode']=df['Age'].fillna(Mod[2])
df
```
